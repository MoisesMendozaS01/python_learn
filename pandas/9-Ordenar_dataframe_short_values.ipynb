```python
import pandas as pd
df_familia = pd.read_csv('familia.csv')
df_familia
```

# 1 Ordenar un dataframe

```python
# ordenar poruna columna
df_familia.sort_values('Edad')
df_familia.sort_values(by='Edad')
```

```python
# Ordenar descendiente por una columna
df_familia.sort_values('Edad',ascending=False)
```

```python
#orden descendiente por multiples columnas
df_familia.sort_values(['Edad','Años Trabajando'],ascending=False)
```

```python
# orden descendiente por multiples columnas y actualizar dataframe
df_familia.sort_values(['Edad','Años Trabajando'],ascending=False,inplace=True)
df_familia
```

```python
# Orden descendiente con una función key
df_familia.sort_values('Nombres',ascending=True,key=lambda col:col.str.lower())
```
